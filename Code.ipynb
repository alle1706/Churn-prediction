```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
from sklearn.metrics import confusion_matrix , classification_report
```

## Loading data and preprocessing

```python
data = pd.read_csv("Churn_Modelling.csv")
data.head()
```

```python
data.drop('CustomerId', axis='columns', inplace=True)
data.drop('Surname', axis='columns', inplace=True)
data.drop('RowNumber', axis='columns', inplace=True)
```

```python
data.drop('Geography', axis='columns', inplace=True)
data.dtypes
```

### Scaling/encoding data

```python
data.head()
```

```python
data['Gender'].replace({'Female':1, 'Male':0}, inplace=True)
```

```python
data.sample(5)
```

```python
cols_to_scale = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']

scaler = MinMaxScaler()
data[cols_to_scale] = scaler.fit_transform(data[cols_to_scale])
```

```python
data.sample(5)
```

## Data split, building the ANN model

```python
x = data.drop('Exited', axis='columns')
y = data['Exited']

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=5)
```

```python
model = keras.Sequential([
    keras.layers.Dense(26, input_shape=(9,), activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=100)
```

### Evaluation

```python
model.evaluate(x_test, y_test)
```

```python
yp = model.predict(x_test)
yp[:5]
```

```python
y_pred = []
for element in yp:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)

y_pred[:10]
```

```python
y_test[:10]
```

```python
print(classification_report(y_test,y_pred))
```
